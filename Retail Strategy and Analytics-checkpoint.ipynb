# Load Required Libraries and Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the datasets
transactionData = pd.read_excel('QVI_transaction_data.xlsx')
customerData = pd.read_csv('QVI_purchase_behaviour.csv')
```

## Exploratory data analysis
<p>The first step in any analysis is to first understand the data.</p>

```python
# Examine transaction data
# Checking data types and sample data
transactionData.info()
transactionData.head()
```

```python
#We can see that the date column is in an integer format.
#Let's change this to a date format.
transactionData['DATE'] = pd.to_datetime(transactionData['DATE'], unit='D')
```

```python
# We should check that we are looking at the right products by examining PROD_NAME.

# Summary of PROD_NAME
transactionData['PROD_NAME'].describe()
```

<p>Looks like we are definitely looking at potato chips but how can we check that
these are all chips? We can do some basic text analysis by summarising the
individual words in the product name.</p>

```python
## Examine the words in PROD_NAME to see if there are any incorrect entries, such as products that are not chips
productWords = transactionData['PROD_NAME'].str.split(expand=True).stack().value_counts().reset_index()
productWords.columns = ['words', 'frequency']

# Remove digits and special characters from words
productWords = productWords[~productWords['words'].str.contains(r'[^A-za-z]', regex=True)]

# Sort the distinct words by frequency of occurence 
productWords.sort_values(by='frequency', ascending=False, inplace=True)
```

<p><b>There are salsa products in the dataset but we are only interested in the chips
    category, so let's remove these.</b></p>

```python
# Remove salsa products
transactionData = transactionData[~transactionData['PROD_NAME'].str.contains('salsa', case=False)]
```

<p><b>Next, we can use `summary()` to check summary statistics such as mean, min and max
values for each feature to see if there are any obvious outliers in the data and if
there are any nulls in any of the columns (`NA's : number of nulls` will appear in
the output if there are any nulls).
</b></p>

```python
# Summarise the data to check for nulls and possible outliers
transactionData.describe()
```

<p><b>There are no nulls in the columns but product quantity appears to have an outlier
which we should investigate further. Let's investigate further the case where 200
packets of chips are bought in one transaction.
</b></p>

```python
##  Filter the dataset to find the outlier

# Find and examine the outlier
outlier_transaction = transactionData[transactionData['PROD_QTY'] == 200]
outlier_transaction
```

<p><b>There are two transactions where 200 packets of chips are bought in one transaction
and both of these transactions were by the same customer</b></p>

```python
## Let's see if the customer has had other transactions

# Check other transactions by the same customer
other_transactions_by_customer = transactionData[transactionData['LYLTY_CARD_NBR'] == 
                                                outlier_transaction['LYLTY_CARD_NBR'].values[0]]

other_transactions_by_customer
```

<p>It looks like this customer has only had the two transactions over the year and is
not an ordinary retail customer. The customer might be buying chips for commercial
purposes instead. We'll remove this loyalty card number from further analysis.
</p>

```python
## Filter out the customer based on the loyalty card number
transactionData = transactionData[transactionData['LYLTY_CARD_NBR'] != outlier_transaction['LYLTY_CARD_NBR'].values[0]]
```

```python
## Re-examine transaction data
transactionData.head()
```

<p><b>That's better. Now, let's look at the number of transaction lines over time to see
if there are any obvious data issues such as missing data.
</b></p>

```python
## Count the number of transactions by date
transaction_count_by_date = transactionData.groupby('DATE').size().reset_index(name='transaction_count')
```

<p><b>There's only 364 rows, meaning only 364 dates which indicates a missing date. Let's
create a sequence of dates from 1 Jul 2018 to 30 Jun 2019 and use this to create a
chart of number of transactions over time to find the missing date.
</b></p>

```python
## Create a sequence of dates and join this the count of transactions by date
## create a column of dates that includes every day from 1 Jul 2018 to
## 30 Jun 2019, and join it onto the data to fill in the missing day.


# Create a sequence of dates fro 1 Jul 2018 to 30 Jun 2019
date_range = pd.date_range(start='2018-07-01', end='2019-06-30')

# Merge the date range with the transaction count data to fill in missing dates
transaction_count_by_date = pd.merge(date_range.to_frame(name='DATE'), transaction_count_by_date,
                                    on='DATE', how='left')

# Fill missing values with 0
transaction_count_by_date['transaction_count'].fillna(0, inplace=True)

# Plot transactions over time
plt.figure(figsize=(12, 6))
plt.plot(transaction_count_by_date['DATE'], transaction_count_by_date['transaction_count'])
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.title('Transactions Over Time')
plt.xticks(rotation=45)
plt.show()
```

<p><b>We can see that there is an increase in purchases in December and a break in late
    December. Let's zoom in on this.</b></p>

```python
# Zoom in on December
december_data = transaction_count_by_date[(transaction_count_by_date['DATE'] >= '2018-12-01') & 
                                          (transaction_count_by_date['DATE'] <= '2018-12-31')]
## Filter to December and look at individual days

# Plot transactions in December
plt.figure(figsize=(12,6))
plt.plot(december_data['DATE'], december_data['transaction_count'])
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.title('Transactions in December 2018')
plt.xticks(rotation=45)
plt.show()
```

<p><b>We can see that the increase in sales occurs in the lead-up to Christmas and that
there are zero sales on Christmas day itself. This is due to shops being closed on
Christmas day.
Now that we are satisfied that the data no longer has outliers, we can move on to
creating other features such as brand of chips or pack size from PROD_NAME. We will
start with pack size.</b></p>

```python
# Create a new column 'PACK_SIZE' by extracting pack size from 'PROD_NAME'
transactionData = transactionData.copy()
transactionData['PACK_SIZE'] = transactionData['PROD_NAME'].str.extract('(\d+)').astype(float)

# Check pack size statistics
pack_size_summary = transactionData['PACK_SIZE'].describe()
print(pack_size_summary) ## The largest size is 380g and the smallest size is 70g - seems sensible!
```

<p><b>Let's plot a histogram of PACK_SIZE since we know that it is a categorical
variable and not a continuous variable even though it is numeric.
</b></p>

```python
## Plot a histogram of pack size
plt.figure(figsize=(10, 6))
sns.histplot(data=transactionData, x='PACK_SIZE', bins=30, kde=True)
plt.xlabel('Pack Size (g)')
plt.ylabel('Frequency')
plt.title('Distribution of Pack Sizes')
plt.show()
```

<p><b>Pack sizes created look reasonable.
Now to create brands, we can use the first word in PROD_NAME to work out the brand
name...</b></p>

```python
# Create a new column 'BRAND' by extracting the brand from 'PROD_NAME'
transactionData['BRAND'] = transactionData['PROD_NAME'].str.split().str[0]
```

<p><b>Some of the brand names look like they are of the same brands - such as RED and
RRD, which are both Red Rock Deli chips. Let's combine these together.</b></p>

```python
# Check unique brand names
unique_brands = transactionData['BRAND'].unique()
print(unique_brands)

# Combine similar brand names
transactionData['BRAND'].replace({'RED': 'RRD', 'INFZNS': 'INFUZIONS', 'SNBTS':'SUNBITES'}, inplace=True)

# Check unique brand names again
unique_brands = transactionData['BRAND'].unique()
print(unique_brands)
```

<p>Let's start with calculating total sales by LIFESTAGE and PREMIUM_CUSTOMER and

```python
## Total sales by LIFESTAGE and PREMIUM_CUSTOMER
total_sales_by_segment = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum().reset_index()

# Calculate the summary of sales by those dimensions and create a plot.
plt.figure(figsize=(12, 6))
sns.barplot(data=total_sales_by_segment, y="LIFESTAGE", x="TOT_SALES", hue='PREMIUM_CUSTOMER', errorbar=None)
plt.xlabel('Total_sales')
plt.ylabel('Customer Segment')
plt.title('Total Sales by Customer Segment')
plt.xticks(rotation=45)
plt.show()
```

Let's see if the higher sales are due to there being more customers who buy chips.</b></p>

```python
# Number of customers by LIFESTAGE and PREMIUM_CUSTOMER
customer_count_by_segment = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['LYLTY_CARD_NBR'].nunique().reset_index()

# Create a bar plot to visualize the number of customers by customer segment
plt.figure(figsize=(12, 6))
sns.barplot(data=customer_count_by_segment, y='LIFESTAGE', x='LYLTY_CARD_NBR', hue='PREMIUM_CUSTOMER', errorbar=None)
plt.xlabel('Number of Customers')
plt.ylabel('Customer Segment')
plt.title('Number of Customers by Customer Segement')
plt.xticks(rotation=90)
plt.show()
```

<p><b>There are more Mainstream - young singles/couples and Mainstream - retirees who buy
chips. This contributes to there being more sales to these customer segments but
this is not a major driver for the Budget - Older families segment.
Higher sales may also be driven by more units of chips being bought per customer. 
    </b></p>

```python
data
```

```python
## Average number of units per customer by LIFESTAGE and PREMIUM_CUSTOMER

# Extract the unit price data for each segment
data['PRICE_PER_UNIT'] = data['TOT_SALES'] / data['PROD_QTY']

# Calculate and plot the average number of units per customer by those two dimensions.
avg_price_per_unit = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PRICE_PER_UNIT'].mean().reset_index()

# Create a bar plot to visualize the average price per unit per by customer segement
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_price_per_unit, x = 'PRICE_PER_UNIT', y='LIFESTAGE', hue='PREMIUM_CUSTOMER', errorbar=None)
plt.xlabel('Average Price per unit')
plt.ylabel('Custoer Segment')
plt.title('Average Price per Unit by Customer Segment')
plt.xticks(rotation=90)
plt.show()
```

<p>Mainstream midage and young singles and couples are more willing to pay more per
packet of chips compared to their budget and premium counterparts. This may be due
to premium shoppers being more likely to buy healthy snacks and when they buy
chips, this is mainly for entertainment purposes rather than their own consumption.
This is also supported by there being fewer premium midage and young singles and
couples buying chips compared to their mainstream counterparts.
</p>

#### Perform an independent t-test between mainstream vs premium and budget midage and young singles and couples

```python
# Perform a t-test to see if the difference is significant.
from scipy.stats import ttest_ind

mainstream_young_midage = data[((data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES') | 
                               (data['LIFESTAGE'] == 'MIDAGE SINGLES/COUPLES')) & (data['PREMIUM_CUSTOMER'] == 'Mainstream')]
budget_premium_young_midage = data[((data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES') |
                                   (data['LIFESTAGE'] == 'MIDAGE SINGLES/COUPLES')) & ((data['PREMIUM_CUSTOMER'] == 'Budget') |
                                  (data['PREMIUM_CUSTOMER'] == 'Premium'))]
t_stat, p_value = ttest_ind(mainstream_young_midage['PRICE_PER_UNIT'], budget_premium_young_midage['PRICE_PER_UNIT'],
                           equal_var=False)
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')

# Check the p-value to determine significance
alpha = 0.05  # Set your significance level
# The t-test results indicate whether there is a significant difference in unit price between the groups
if p_value < alpha:
    print("The unit price for mainstream, young and mid-age singles and couples is significantly higher than"
+" that of budget or premium, young and midage singles and couples.")
else:
    print('There is no significant differencein unit price between the groups.')
```

## Deep dive into specific customer segments for insights
<p>We have found quite a few interesting insights that we can dive deeper into.
We might want to target customer segments that contribute the most to sales to
retain them or further increase sales. Let's look at Mainstream - young
singles/couples. For instance, let's find out if they tend to buy a particular
brand of chips.</p>

```python
## Deep dive into Mainstream, young singles/couples
## Work out of there are brands that these two customer segments prefer more than others. 
## You could use a technique called affinity analysis or a-prior analysis

# Filter the DataFrame to select only the Mainstream - young singles/couples segment
target_segment = data[((data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES') | 
                               (data['LIFESTAGE'] == 'MIDAGE SINGLES/COUPLES')) & (data['PREMIUM_CUSTOMER'] == 'Mainstream')]
# Analyze preferred brands
preferred_brands = target_segment['BRAND'].value_counts()

# Plot preferred brands
plt.figure(figsize=(10, 6))
preferred_brands.plot(kind='bar')
plt.title('Preferred Brands for Mainstream - Young Singles/Couples')
plt.xlabel('Brand')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

## Preferred Brands(We can see that) :
<li><b>Kittle Brand</b> appears to be the most preferred barnd among Mainstream - young singles/couples, as it has
the highest number of purchases within this segement</li>
<li><b>Pringles Brand, Doritos Brand and Smiths Brand</b> also have a significant number of purchases but are less
    preferred compared to the <b>Kittle Brand</b></li>
<li>The Mainstream - young singles/couples segment seems to have clear preference for the Kittle brand over other brands</li>

```python
# Analyze preferred pack sizes
preferred_pack_sizes = target_segment['PACK_SIZE'].value_counts()

# Plot preferred pack sizes
plt.figure(figsize=(10, 6))
preferred_pack_sizes.plot(kind='bar')
plt.title('Preferred Pack Sizes for Mainstream - Young Singles/Couples')
plt.xlabel('Pack Size (g)')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

## Preferred Pack Sizes (We can see that):
<li>Mainstream - young singles/couples tend to prefer <b>pack sizes around 175g</b> as it has the highest number
of purchases in this segment.</li>
<li><b>Pack sizes around 15g</b> are second most preferred among this segment.</li>
<li>Larger pack sizes (e.g., 200g) are less preferred among this segment</li>
